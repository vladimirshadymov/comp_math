### Решим уравнение аналитически

Ищем решение уравнения в виде $x(t)=t^\alpha$, тогда уравнение примет вид:
$$\alpha(\alpha-1)\cdot t^{\alpha-2} + 2\alpha\cdot t^{\alpha -1}+2.5\cdot t^{\alpha -2}=0,$$
$$(\alpha^2+\alpha+2.5)t^{\alpha -2}=0,$$
т.к. $t\neq 0$, то уравнение выше эквивалентно
$$\alpha^2+\alpha+2.5=0,$$
откуда 
$$\alpha = -\frac{1}{2}\pm\frac{3}{2}i,$$
то есть комплексное частное решение исходного однородного уравнения
$$x_o(t)=t^{-\frac{1}{2}\pm\frac{3}{2}i},$$
или
$$x_o(t)=\frac{e^{\pm\frac{3}{2}i}}{\sqrt{t}}$$
Его мнимая и действительная части линейно независимы, поэтому общее действительное решение имеет вид
$$x(t)=C_1\frac{\sin\left(\frac{3}{2}\ln t\right)}{\sqrt{t}}+C_2\frac{\cos\left(\frac{3}{2}\ln t\right)}{\sqrt{t}}.$$
Теперь решим задачу Коши, найдем коэффициенты $C_1$ и $C_2$ из начальных условий $x(1)=1,\ \dot x(1)=0$
$$x(t)|_{t=1} = C_1\frac{\sin\left(\frac{3}{2}\ln t\right)}{\sqrt{t}}+C_2\frac{\cos\left(\frac{3}{2}\ln t\right)}{\sqrt{t}}\bigg|_{t=1}=C_2=1$$
$$\dot x(t)|_{t=1} = 
\frac{(3C_1-C_2)\cos\left(\frac{3}{2}\ln t\right)+
(-3C_2-C_1)\sin\left(\frac{3}{2}\ln t\right)}
{2\cdot t^{\frac{3}{2}}}\bigg|_{t=1}=
(3C_1-C_2)=0$$
Тогда решение задачи Коши:
$$x(t)=\frac{\sin\left(\frac{3}{2}\ln t\right)}{3\sqrt{t}}+\frac{\cos\left(\frac{3}{2}\ln t\right)}{\sqrt{t}}$$

```python
# функция аналитического решения задачи Коши
def get_exact_x(t):
    return (np.sin(1.5*np.log(t))/t**0.5/3+np.cos(1.5*np.log(t))/t**0.5)
```

Построим график аналитического решения

```python
t_values = np.linspace(1, 101, 500) # диапазон точек
x_values_exact_sol = get_exact_x(t_values)
plt.plot(t_values, x_values_exact_sol)
plt.title("Аналитическое решение")
plt.xlabel('$t$')
plt.ylabel('$x(t)$')
plt.grid()
plt.show()
```

### Решим уравнение численно

Для того чтобы численно решить уравнение Эйлера перепишем его в виде системы д.у.
$$
\begin{cases}
    \dot{x} = p \\
    \dot{p} = -2\frac{p}{t}-2.5\frac{x}{t^2}
\end{cases}
$$
т.е система вида 
$$\dot{\overline x} = F(t, \overline x)$$
Тогда для исходной системы можно задать следующим образом

```python
def F(x_values):
    res = []
    t = x_values[0]
    x1 = x_values[1]
    x2 = x_values[2]
    res.append(1.)
    res.append(x2)
    res.append(-2*x2/t - 2.5*x1/t**2)
    return np.array(res)
```

### Явный метод Эйлера
$$\frac{\overline x_{n+1}-\overline x_n}{\tau}=F(t_n, \overline x_n)$$

```python
def euler_explicit(F, x_start, t_stop, t_start=1., tau=0.1):
    """
    Явный метод Эйлера
    :param F: функция системы
    :param x_start: условие Коши, включает в себя значение функций и ее производных
    :param t_start: начальное время
    :param t_stop: конечное время
    :param tau: временной шаг
    :return: значения функции в соотв. точках
    """
    # Начальное значение
    x = x_start.copy()
    res = []
    while x[0] < t_stop:
        x += tau*F(x)
        res.append(x.copy())
    return np.array(res).T
```

```python
def plot_solution(method):
    
    T_values = [(1+10**i) for i in range(4)]
    fig, axs = plt.subplots(len(T_values)+1)

    for i in range(len(T_values)):
        T = T_values[i]
        # точное решение
        t_values = np.linspace(1., T, 500)
        x_exact = get_exact_x(t_values)
        axs[i].plot(t_values, x_exact, label=r'Аналитическое решение')
    
    
        x_init = np.array([1., 1., 0.])
        # Явный Эйлер tau=0.1
        res = method(F, x_start=x_init, t_stop=T)
        t_values = res[0].copy()
        x_values_euler = res[1].copy()
        axs[i].plot(t_values, x_values_euler, label=r'$\tau=0.1$')
        error01 = get_exact_x(t_values)-x_values_euler
        error01 = np.abs(error01)
        t_values01 = t_values.copy()
    
        # Явный Эйлер tau=0.05
        res = method(F, x_start=x_init, t_stop=T, tau=0.05)
        t_values = res[0].copy()
        x_values_euler = res[1].copy()
        axs[i].plot(t_values, x_values_euler, label=r'$\tau=0.05$')
        error005 = get_exact_x(t_values)-x_values_euler
        error005 = np.abs(error005)
        t_values005 = t_values.copy()
    
        # Явный Эйлер tau=0.01
        res = method(F, x_start=x_init, t_stop=T, tau=0.01)
        t_values = res[0].copy()
        x_values_euler = res[1].copy()
        axs[i].plot(t_values, x_values_euler, label=r'$\tau=0.01$')
        error001 = get_exact_x(t_values)-x_values_euler
        error001 = np.abs(error001)
        t_values001 = t_values.copy()
        axs[i].grid()
        axs[i].set_title('T=%d' % T)
        axs[i].set_xlabel('$t$')
        axs[i].set_ylabel('$x(t)$')
        axs[i].legend()
    
    axs[-1].plot(t_values01, error01, label=r'$\tau=0.1$')
    axs[-1].plot(t_values005, error005, label=r'$\tau=0.05$')
    axs[-1].plot(t_values001, error001, label=r'$\tau=0.01$')
    axs[-1].set_title("Зависимость невязки от времени")
    axs[-1].legend()
    axs[-1].grid()

    fig.set_tight_layout(tight=8)
    fig.set_figwidth(12)
    fig.set_figheight(30)
    plt.show()
```

```python
plot_solution(euler_explicit)
```

### Неявный метод Эйлера
$$\frac{\overline x_{n+1}-\overline x_n}{\tau}=F(t_{n+1}, \overline x_{n+1})$$

```python
def euler_implicit(F, x_start, t_stop, t_start=1., tau=0.1):
    """
    Неявный метод Эйлера с предиктором и корректором
    """
    # Начальное значение
    x = x_start.copy()
    res = []
    while x[0] < t_stop:
        # Прогноззируем f_i+1
        x_tmp1 = x + tau*F(x)
        # Корректируем
        x_tmp2 = x + tau*(F(x) + F(x_tmp1))/2
        x += tau*(F(x) + F(x_tmp2))/2
        res.append(x.copy())
    return np.array(res).T
```

```python
plot_solution(euler_implicit)
```

### Метод Эйлера с центральной точкой
$$\frac{\overline x_{n+1}-\overline x_{n-1}}{2\tau}=F(t_n, \overline x_n)$$
или же, если расписать по итерациям (выполняется только при равномерной сетке)
$$
\overline x_{n+1/2} = \overline x_n + \frac{\tau}{2}F(t_n, \overline x_n)
$$
$$
\overline x_{n+1} = \overline x_n + \tau F\left(t_n+\frac{\tau}{2}, \overline x_{n+1/2}\right)
$$

```python
def euler_central_point(F, x_start, t_stop, t_start=1., tau=0.1):
    """
    Метод Эйлера с центральной точкой
    """
    # Начальное значение
    x = x_start.copy()
    res = []
    while x[0] < t_stop:
        # Прогнозируем f_(i+1/2), а затем взять полусумму с f_i
        x_tmp = x + tau/2*F(x)
        x += tau*F(x_tmp)
        res.append(x.copy())
    return np.array(res).T
```

```python
plot_solution(euler_central_point)
```

### Метод трапеций
$$
x_{n+1}^* = \overline x_n + \tau F(t_n, \overline x_n), \\
\overline x_{n+1} = \overline x_n + \frac{\tau}{2} (F\left(t_n, \overline x_{n}\right) + F\left(t_{n+1}, \overline x_{n+1}^*\right))
$$

```python
def trapezium(F, x_start, t_stop, t_start=1., tau=0.1):
    """
    Метод трапеций
    """
    # Начальное значение
    x = x_start.copy()
    res = []
    while x[0] < t_stop:
        # Прогнозируем f_(i+1), а затем взять полусумму с f_i
        x_tmp = (F(x + tau*F(x)) + F(x))/2
        # Результат прибавлять в качестве вектора касательной
        x += tau*x_tmp
        res.append(x.copy())
    return np.array(res).T
```

```python
plot_solution(trapezium)
```

### Методы Рунге-Кутты

Теперь решим исходное уравнение, используя семейство методов Рунге-Кутты, разных порядков. Пользоваться будем классическими методами Рунге-Кутты (к вопросу о выборе коэфициентов).

$$
k_1 = F(t_n, x_n),\\
k_2 = F(t_n + \alpha_2 \tau, x_n +\tau \beta_{21}k_1), ... ,\\
k_r = F(t_n + \alpha_r \tau, x_n + \tau (\beta_{r1}k_1+...+\beta_{rr-1}k_r))\\
x_{n+1} = x_n + \tau(\gamma_1k_1+...+\gamma_rk_r)
$$
где $\alpha_p$, $\beta_{ij}$, $\gamma_s$ - значения из таблицы Бутчера.

Запишем таблицы Бутчера для разных порядков аппроксимации

```python
#таблицы Бутчера
butcher_matrices = [None]

# 1-й порядок аппрокимации
tmp = [[0,0],
       [0,1.]]
butcher_matrices.append(tmp)

# 2-й порядок аппрокимации
tmp = [[0.,0, 0],
       [0.5,0.5, 0],
       [0, 0., 1.]]
butcher_matrices.append(tmp)

# 3-й порядок аппрокимации
tmp = [[0., 0, 0, 0],
       [0.5, 0.5, 0, 0],
       [1., 0., 1., 0],
       [0, 1./6., 2./3., 1./6.]]
butcher_matrices.append(tmp)

# 4-й порядок аппрокимации
tmp = [[0., 0, 0, 0, 0],
       [1./3., 1./3., 0, 0, 0],
       [2./3., -1./3., 1., 0, 0],
       [1., 1., 1., 1., 0],
       [0, 1./8., 3./8., 3./8., 1./8.]]
butcher_matrices.append(tmp)
```

```python
def rune_kutta_iter(F, x_init, deg=4, tau=0.1):
    butcher_matrix = butcher_matrices[deg].copy()
    k = [np.zeros_like(x_init) for i in range(len(butcher_matrix)-1)]
    r = len(butcher_matrix)-1
    for i in range(r):
        k_tmp = np.zeros_like(x_init)
        for j in range(1,len(butcher_matrix[i])):
            k_tmp += butcher_matrix[i][j]*k[j-1]
        k_tmp[0] = butcher_matrix[i][0]*x_init[0]
        k[i] = F(x_init+tau*k_tmp)
    k_shift = np.zeros_like(x_init)
    for i in range(r):
        k_shift += butcher_matrix[r][i+1]*k[i] 
    x_next = x_init + tau*k_shift
    
    return np.array(x_next)
    

def method_rune_kutta(F, x_start, t_stop, t_start=1., deg=4, tau=0.1):
    '''
    Метод Рунге-Кутты
    '''
    # Начальное значение
    x = x_start.copy()
    res = []
    while x[0] < t_stop:
        res.append(x.copy())
        x = rune_kutta_iter(F, x, deg=deg, tau=tau)
    return np.array(res).T
```

```python
def plot_rune_kutta(deg):
    T_values = [(1+10**i) for i in range(4)]
    fig, axs = plt.subplots(len(T_values)+1)

    for i in range(len(T_values)):
        T = T_values[i]
        # точное решение
        t_values = np.linspace(1., T, 500)
        x_exact = get_exact_x(t_values)
        axs[i].plot(t_values, x_exact, label=r'Аналитическое решение')
    
    
        x_init = np.array([1., 1., 0.])
        # Метод Рунге-Кутты tau=0.1
        res = method_rune_kutta(F, x_start=x_init, t_stop=T, t_start=1., deg=deg, tau=0.1)
        t_values = res[0].copy()
        x_values_euler = res[1].copy()
        axs[i].plot(t_values, x_values_euler, label=r'$\tau=0.1$')
        error01 = get_exact_x(t_values)-x_values_euler
        error01 = np.abs(error01)
        t_values01 = t_values.copy()
    
        # Метод Рунге-Кутты tau=0.05
        res = method_rune_kutta(F, x_start=x_init, t_stop=T, t_start=1., deg=deg, tau=0.05)
        t_values = res[0].copy()
        x_values_euler = res[1].copy()
        axs[i].plot(t_values, x_values_euler, label=r'$\tau=0.05$')
        error005 = get_exact_x(t_values)-x_values_euler
        error005 = np.abs(error005)
        t_values005 = t_values.copy()
    
        # Метод Рунге-Кутты tau=0.01
        res = method_rune_kutta(F, x_start=x_init, t_stop=T, t_start=1., deg=deg, tau=0.01)
        t_values = res[0].copy()
        x_values_euler = res[1].copy()
        axs[i].plot(t_values, x_values_euler, label=r'$\tau=0.01$')
        error001 = get_exact_x(t_values)-x_values_euler
        error001 = np.abs(error001)
        t_values001 = t_values.copy()
        axs[i].grid()
        axs[i].set_title('T=%d' % T)
        axs[i].set_xlabel('$t$')
        axs[i].set_ylabel('$x(t)$')
        axs[i].legend()
    
    axs[-1].plot(t_values01, error01, label=r'$\tau=0.1$')
    axs[-1].plot(t_values005, error005, label=r'$\tau=0.05$')
    axs[-1].plot(t_values001, error001, label=r'$\tau=0.01$')
    axs[-1].set_title("Зависимость невязки от времени")
    axs[-1].legend()
    axs[-1].grid()

   
    fig.set_tight_layout(tight=10)
    fig.set_figwidth(12)
    fig.set_figheight(30)
    plt.show()
```

### №2
Получить траекторию движения спутника вокруг планеты, проведя численное решение задачи двух тел 


$$
\begin{cases}
    \dot{x} = z \\
    \dot{y} = u \\
    \dot{z} = -\frac{x}{(x^2+y^2)^{3/2}} \\
    \dot{u} = -\frac{x}{(x^2+y^2)^{3/2}} \\
\end{cases}
$$

при $t \in [0, 20]$, с начальными условиями $x(0)=0.5,\ y(0)=0,\ z(0)=0,\ u(0)=\sqrt{3}$. 

Запишем функцию системы:

```python
# функция системы
def gravity_F(p):
    t = p[0]
    x = p[1]
    y = p[2]
    z = p[3]
    u = p[4]
    res = []
    res.append(1.)
    res.append(z)
    res.append(u)
    res.append(-x/(x**2+y**2)**1.5)
    res.append(-y/(x**2+y**2)**1.5)
    
    return np.array(res)
```

Найдем численно траектории для разных порядков аппроксимации и разных шагов численного интегрирования системы

```python
degs = [1, 2, 3, 4]
tau_values = np.linspace(1e-3, 1e-1, 6)
fig, axs = plt.subplots(4)

t_stop = 20.
t_start = 0.
p_init = [0., 0.5, 0., 0., 3**0.5]

for deg in degs:
    for tau in tau_values:
        res = method_rune_kutta(gravity_F, x_start=p_init, t_stop=t_stop, t_start=t_start, deg=deg, tau=tau)
        axs[deg-1].plot(res[1], res[2], label=r'$\tau=%f$'%tau)
    axs[deg-1].grid()
    axs[deg-1].set_title('Степень аппроксимации метода Рунге-Кутте %d'%deg)
    axs[deg-1].legend()
    axs[deg-1].set_xlim(-2.5,+2.5)
    axs[deg-1].set_ylim(-2,+2)
    axs[deg-1].set_xlabel('$x(t)$')
    axs[deg-1].set_ylabel('$y(t)$')

fig.set_tight_layout(tight=10)
fig.set_figwidth(12)
fig.set_figheight(25)
plt.show()
```

__Выводы:__ видно, что начиная с некоторого $\tau$,  1-я и 4-я степень аппроксимации дают неустойчивые решения. 1-я это метод Эйлера. Он не сохраняет энергию. Поэтому даже при небольших $\tau$ решения начинают расходится. 4-я же степень наоборот удаляет энергию из системы, т.к. происходит избыточная оценка вектора смещения в ходе вычисления слдующей итерации.

### №3

Дано уравнение 
$$\ddot x +\sin x=0$$
Перепишем в виде системы
$$
\begin{cases}
    \dot x = v \\
    \dot v = -\sin x
\end{cases}
$$
Линеаризованная система:
$$
\begin{cases}
    \dot x = v \\
    \dot v = -x
\end{cases}
$$
Тогда особая точка $x=0,\ \ v=0$
Откуда собственные зачения для матрицы правой части $\lambda = \pm i$,
то есть $(0, 0)$ - особая точка центр.

Тогда собственные вектора 
$$
\overline x_{1,2} = (0\ \  1)^T \pm i(1\ \ 0)^T
$$

тогда векторы $(0\ \  1)^T ,\ \  (1\ \ 0)^T$ - направляющие полуосей центра.
Теперь численно решим систему

```python
def sin_vib(p):
    t = p[0]
    x = p[1]
    v = p[2]
    res = []
    res.append(1.)
    res.append(v)
    res.append(-np.sin(x))
    return np.array(res)
```

```python
deg = 3
tau_values = np.linspace(1e-3, 1e-1, 6)
fig, axs = plt.subplots(2)
t_stop = 100.
t_start = 0.
p_init = np.array([0., 0., .9])

for tau in tau_values:
    res = method_rune_kutta(sin_vib, x_start=p_init, t_stop=t_stop, t_start=t_start, deg=deg, tau=tau)
    axs[0].plot(res[1], res[2], label=r'$\tau=%f$'%tau)    
    axs[1].plot(res[0], res[1], label=r'$\tau=%f$'%tau)
    
axs[0].set_xlabel(r'$x(t)$')
axs[0].set_ylabel(r'$v(t)$')
axs[1].set_xlabel(r'$t$')
axs[1].set_ylabel(r'$x(t)$')
axs[0].legend()
axs[1].legend()
axs[0].grid()
axs[1].grid()
axs[0].set_title(r'Фазовые траектории')
axs[1].set_title(r'Координата от времени')

fig.set_tight_layout(tight=10)
fig.set_figwidth(12)
fig.set_figheight(12)
plt.show()
```
